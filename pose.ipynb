```python
import cv2
import mediapipe as mp
import math
```

```python
def calculateAngle(landmark1, landmark2, landmark3):
    # 獲取所需座標
    x1, y1, _ = landmark1
    x2, y2, _ = landmark2
    x3, y3, _ = landmark3
    # 計算三點之間的夾角
    angle = math.floor(
        math.degrees(math.atan2(y3 - y2, x3 - x2) - math.atan2(y1 - y2, x1 - x2))
    )
    if angle < 0: # 負角度 可理解為在第四象限-60度 等於順時針60度 等於負時針300度 
        angle += 360 
    if angle > 180: # 大角換小角  
        angle = 360 - angle
    return angle
```
